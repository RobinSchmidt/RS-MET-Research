```python
f(x) = 1 / (1 + x^2)   # = sum_n  x^2n    ?
g(x) = 1 / (1 - x^2)   # = sum_n (-x^2)^n ?
f, g
```

```python
# Let sage compute the Taylor series expansion of 
# order 10 of both functions:
f.series(x==0, 10), g.series(x==0, 10)
```

```python
# plot both functions together with their Taylor series of
# various orders:
xmin, xmax = -2, 2
pf, pg = Graphics(), Graphics()
N  = 7
for i in range(0, N, 2):  # odd coeffs are zero due to symmetry
    pf += plot(taylor(f(x), x, 0, i), xmin, xmax, color="gray")
for i in range(0, N, 2):  # odd coeffs are zero due to symmetry
    pg += plot(taylor(g(x), x, 0, i), xmin, xmax, color="gray")
pf += plot(f, xmin, xmax, color="black")
pg += plot(g, xmin, xmax, color="black", detect_poles="show")    
graphics_array([[pf, pg]]).show(frame=True, axes=False, figsize=[10,4],
                               ymin=-1,ymax=2)
```

```python
# plot contour plots for both functions in the complex plane
var("z")
#z = x + I*y
#fc(x,y) = f(x + I*y) # lowercase i is not available anmore - because of using it for the loop?
fc = lambda x,y: log(abs(f(x+I*y)))
gc = lambda x,y: log(abs(g(x+I*y)))
pf = contour_plot(fc, [-2,2], [-2,2], labels=True, label_colors='white')
pg = contour_plot(gc, [-2,2], [-2,2], labels=True, label_colors='white')
graphics_array([[pf, pg]]).show(frame=True, axes=False, figsize=[10,4],
                                xmin=-2,xmax=2,ymin=-2,ymax=2)
# it takes very long to compute the plot - maybe it's because w use
# symbolic functions?
```

The function f has poles at i,-i whereas the function g has poles at 1,-1. Both has the same radius of convergence around 0 because this radius is determined by the distance to the nearest pole.
